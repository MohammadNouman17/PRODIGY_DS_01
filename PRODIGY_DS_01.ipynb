```python
final_ds=df.dropna()
final_ds
```

```python
# Distribution for a continuous variable

sns.histplot(final_ds['Age'])
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sex_count = pd.value_counts(df['Sex'])
print(sex_count)
```

```python
# Distribution of a categorical variable Gender.

plt.bar(sex_count.index, sex_count.values)
plt.title('Distribution of Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```
